```python
# define the data
season_data = {
    'Season': [1, 2, 3, 4, 5],
    'Avg. Viewers (millions)': [1.23, 1.30, 1.52, 1.87, 4.32]
}

# create a df from the data
season_df = pd.DataFrame(season_data)

# display the df 
print(season_df.to_markdown(index=False))
```

```python
# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(season_df['Season'], season_df['Avg. Viewers (millions)'], marker='o', linestyle='-', color='blue')  
plt.title('Average Viewership Over Seasons') 
plt.xlabel('Season')  
plt.ylabel('Avg. Viewers (millions)')  
plt.xticks(season_df['Season'])  
plt.grid(True) 
plt.show()  
```

```python
# calculate the change in viewership
season_df['Viewership Change'] = season_df['Avg. Viewers (millions)'].diff()

# plotting the bar graph for the changes
plt.figure(figsize=(10, 6))
bars = plt.bar(season_df['Season'][1:], season_df['Viewership Change'][1:], color='skyblue')
plt.title('Change in Viewership from Season to Season')
plt.xlabel('Season')
plt.ylabel('Change in Avg. Viewers (millions)')
plt.xticks(season_df['Season']) 
plt.grid(axis='y')

# labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')  # va: vertical alignment

plt.show()
```

```python
# extract specific changes for the description
change_3_to_4 = season_df.loc[season_df['Season'] == 4, 'Viewership Change'].iat[0]
change_4_to_5 = season_df.loc[season_df['Season'] == 5, 'Viewership Change'].iat[0]

# description with inline references to the changes
description = f"""
As observed in the graph, the viewership changes between seasons show varying trends. Notably, the viewership increased by {change_3_to_4:.2f} million from Season 3 to 4, and saw a significant jump by {change_4_to_5:.2f} million from Season 4 to 5.
"""

# display the description 
print(description)
```
